```python
#Method Overloading - child class method will always override 
#the parent class method
#                  core rules:
#The method name must be the same in both classes.
#The parameters must be the same.
#There must be an Inheritance (Parent-Child) relationship.


class parent:
    def a(self):
        print("I am going to work")
class child(parent):
    def a(self):
        print("I am going to school")

c1 = child()
c1.a()
```

```python
#Multiple Inheritance 
#In Multiple Inheritance (when a child has two parents),
#Python needs a rule to decide which parent to look at first if both have a method with the same name. it uses method resolution order(MRO).
class Father:
    def a(self):
        print("hello")
class Mother:
    def b(self):
        print("hi")
class Child(Father, Mother):
    def c(self):
        print("hello hi")

c1 = Child()
c1.a()
c1.b()

```

```python
#Accessing private member from parent class to child class through getter method 
class Parent:
    def __init__(self):
        self.__x = 10
    def getValue(self):
        return self.__x
class Child(Parent):
    def getValue(self):
        return super().getValue()

c1 = Child()
c1.getValue()
```

```python
class Parent:
    def __init__(self, balance):
        self._balance = balance
    def Show_balance(self):
        print("Balance is :", self._balance)
class Child(Parent):
    def deposit(self):
        amount = int(input("Enter amount u want to deposit"))
        self._balance += amount
    def withdraw(self):
        withdraw = int(input("Enter money u want to withdraw:"))
        if withdraw > self._balance :
            print("Insufficient balance ")
        else :
            self._balance -= withdraw
            print("Withdraw Successful ")

ch1 = Child(1233)
ch1.deposit()
ch1.withdraw()
ch1.Show_balance()
```
