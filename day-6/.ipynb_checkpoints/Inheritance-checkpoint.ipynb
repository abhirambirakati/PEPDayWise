```python
#inheritance - derived class uses base class features
class Human:   
    def eat(self):
        print("Human can eat")
class Student(Human):
    def study(self):
        print("Student can study")

s1 = Student()
s1.eat()
s1.study()
```

```python
class Animal:
    def a(self):
        print("Animal can only speak")
class Dog(Animal):
    def c(self):
        print("but Dog only can bark")

d1 = Dog()
d1.c()
d1.a()
```

```python
class Parent:
    def nokia(self):
        print("Nokia mobile ")
    def call(self):
        print("it can only call")
class Child(Parent):
    def Smart(self):
        print("smartPhone")
    def internet(self):
        print("In internet we can access browaing easily")
c1 = Child()
c1.nokia()
c1.call()

c1.Smart()
c1.internet()
```

```python
class Parent:
    def __init__(self):
        print("This is parent constructor")
class Child(Parent):
    def __init__(self):
#super() is used to call methods from the parent class inside the child class.
# this we need to use when when we have functions, if its an constructor we need to use def init like that
        super().__init__()
        print("This is child constructor")
c1 =Child()

```

```python
#Method Overloading - child class method will always override 
#the parent class method
#                  core rules:
#The method name must be the same in both classes.
#The parameters must be the same.
#There must be an Inheritance (Parent-Child) relationship.


class parent:
    def a(self):
        print("I am going to work")
class child(parent):
    def a(self):
        print("I am going to school")

c1 = child()
c1.a()
```

```python
#Multiple Inheritance 
#In Multiple Inheritance (when a child has two parents),
#Python needs a rule to decide which parent to look at first if both have a method with the same name. it uses method resolution order(MRO).
class Father:
    def a(self):
        print("hello")
class Mother:
    def b(self):
        print("hi")
class Child(Father, Mother):
    def c(self):
        print("hello hi")

c1 = Child()
c1.a()
c1.b()

```

```python
class Parent:
    def shape(self):
        print("The shape is triangle")
class Child(Parent):
    def shape(self):
        super().shape() 
        print("The shape is circle")

c1 = Child()
c1.shape()
```

```python
#in which teaching,coding,teacher +coding
class Teacher:
    def a(self):
        print("Teaching")
class Coder:
    def b(self):
        print("Coding")
class Student(Teacher, Coder):
    pass

s1 = Student()
s1.a()
s1.b()
```

```python
class Teacher:
    def ab(self):
        print("Teaching")
class Coder:
    def bb(self):
        print("Coding")
class Student(Teacher, Coder):
    def ca(self):
        self.ab()
        self.bb()

s1 = Student()
s1.ab()
s1.bb()
print("Combined skills :")
s1.ca()
```

```python
class Parent:
    def __init__(self):
        # Private attribute (Name Mangled)
        self.__house_name = "Abhiram's Villa"

    # Public Getter method to allow access
    # def get_house(self):
    #     return self.__house_name

class Child(Parent):
    def show_property(self):
        print(f"I am inheriting: {self.get_house()}")

c1 = Child()
c1.show_property()
```

```python
class Parent:
    def __init__(self):
        self.__house = "Luxury Villa" # Private

class Child(Parent):
    def display(self):
        # Accessing via the mangled name instead of a getter
        print(self._Parent__house)

c1 = Child()
c1.display()
```
