```python
class Human:
    def speak(self):
        print("Human is speaking")
class Baby:
    def speak(self):
        print("Baby is crying")
class Dog:
    def speak(self):
        print("Dog is Barking")

h = Human()
b = Baby()
D = Dog()
h.speak()
b.speak()
D.speak()
```

```python
class Person:
    def role(self):
        print("im person")
class Student(Person):
    def role(self):
        print("Im student")
class Teacher(Person):
    def role(self):
        print("Im Teacher")
class Assistant(Student, Teacher):
    pass

a1 = Assistant()
a1.role()
```
