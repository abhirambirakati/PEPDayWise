```python
c =  int(input("Enter Your Age:"))
if c>=90:
    print("Grade A")
elif c>=75:
    print("Grade B")
elif c>=50:
    print("Grade C")
else:
    print("Fail")
```

```python
#Check which One is bigger
d = int(input("Enter a Integer:"))
e =  int(input("Enter a Integer:"))
if d>e:
    print("D is Bigger :",d)
elif d<e:
    print("E is Bigger Integer :",e)
else:
    print("Both are Equal")
```

```python
ab = str(input("Enter a String:"))
if (ab == "Green"):
    print("You can go")
elif (ab == "yellow"):
    print("Warning")
elif (ab == "red"):
    print("You have to stop:")
else:
    print("Invalid input")
```
