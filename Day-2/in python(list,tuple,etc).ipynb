```python
#Creating a empty dictionary and then fill values by user with
contact = {}
while True:
    print("\n-----Contact Book----")
    print("1.Add Contact")
    print("2.View Contact")
    print("3.Search Contact")
    print("4. Delete Contact")
    print("5 . Exit")

    choice = input("Enter Your Choice:")

    #For Adding Contact
    if choice == "1":
        name = input("Enter Your name:")
        phone = input("Enter Your Number:")
        contact[name] = phone
        print("Contact Added Succesfully")
     #For Viewing Contact   
    elif choice == "2":
        if contact:
            print("\n Saved Contact")
            for name,phone in contact.items():
                print(name, " : " , phone)
        else:
            print("No Contact Found")
    elif choice == "3":
        name = input("Enter the name u want:")
        if name in contact:
            print("Phone Number: " ,contact[name])
        else:
            print("Contact not found:")

     # 4 Delete Contact
    elif choice == "4":
        name = input("Enter the name u want:")
        if name in contact:
            contact.pop(name)
            print("Contact Deleted")
        else:
            print("Contact not found")

    # 5 Exit

    elif choice == "5":
        print("Thnak u for using the contact book")
        break
    
        
        
        
```

```python
Student = {}
while True:
    print("\n--- Student Details ---")
    print("1. Add Student Information")
    print("2. View Saved Students")
    print("3. Search Student (by Roll No)")
    print("4. Delete Student")
    print("5. Exit")

    choice = input("Enter Your Choice: ")

    if choice == "1":
        name = input("Enter Student name: ")
        roll_no = input("Enter Student Roll_no: ")
        course = input("Enter Student Course: ")
        # You created keys as strings "name" and "course"
        Student[roll_no] = {
            "name" : name,
            "course" : course
        }       
        print("Added Successfully")

    elif choice == "2":
        if Student:
            # Note the variables roll and info
            for roll, info in Student.items():
                print(f"\nRoll No: {roll}")
                # You must use quotes around "name" and "course"
                print("Name:", info["name"])
                print("Course:", info["course"])
        else:
            print("No student records found.")
        
    elif choice == "3":
        roll_no = input("Enter roll number u want to search: ")
        if roll_no in Student:
            records = Student[roll_no]
            # Use quotes here as well
            print("Name:", records["name"])
            print("Course:", records["course"])
        else:
            print("Student not found.")

    elif choice == "4":
        roll_no = input("Enter roll number u want to delete: ")
        removed = Student.pop(roll_no, "Not Found")
        if removed == "Not Found":
            print("Roll no not found")
        else:
            print(f"Student {removed['name']} successfully deleted.")

    elif choice == "5":
        print("Thank you for using student record system.")
        break

    else:
        print("Invalid Choice")
```
