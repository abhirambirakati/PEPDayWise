```python
import re
text = "cat cat cut"
result = re.findall("c.t", text)
print(result)
#2.Start of a String
text1 = "Hello World"
print(bool(re.search("^Hello", text1)))
print(bool(re.search("^World", text1)))
#3.End of a String
print(bool(re.search("Hello$", text1)))
print(bool(re.search("World$", text1)))
#4.Star
text2 = "helloooo"
result = re.findall("lo*", text2)
print(result)
#5.one or more(+)
result1 = re.findall("lo+", text2)
print(result1)
#6. zero or one time(?)
text3 = "color colour"
result2 = re.findall("colou?r",text)
print(result2)
#7. Character Set([])
text3 = "apple ball cat"
result3 = re.findall("[abc]", text3)
print(result3)
#8.Digits([0-9])
text4 = "My age is 30"
result4 = re.findall("[0-9]",text4)
print(result4)
#9.([A-Z] Capital letters
text5 = "RAHUL"
resutl5 = re.findall("[A-Z]", text5)
print(resutl5)
#10.([a-z] Capital letters
text6 = "rahul is a goood"
resutl5 = re.findall("[a-z]", text6)
print(resutl5)
```

```python

#11.([A-Z a-z] Capital letters
text7 = "Rahul"
resutl7 = re.findall("[A-Z a-z]", text7)
print(resutl7)
#12.(\d) Digits 
text8 = "Marks: 90"
result8 = re.findall("\d", text8)
print(result8)
#13.("\D") No digits
result9 = re.findall("\D", text8)
print(result9)
#14.
#20.Grouping(()))
text12 = "abab ab"
result132 = re.findall("(ab)+",text12)
print(result132)
```
