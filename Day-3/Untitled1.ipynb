```python
text = "naman"
left = 0
right = len(text) - 1
is_palindrome = True

while left < right:
    if text[left] != text[right]:
        is_palindrome = False
        break
    
    # These must be outside the 'if' to move to the next characters
    left += 1
    right -= 1

print(is_palindrome)
```
