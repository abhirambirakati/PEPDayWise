```python
# 1. Encapsulation = Data Hiding + Control Access
class Student:
    def __init__(self, marks):
        self.__marks = marks
    def get_marks(self):
        return self.__marks
    def set_marks(self,new):
        self.__marks = new
        
s1 = Student(100)
s1.set_marks(99)
print(s1.get_marks())
```

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance
        
    def deposit(self, amount):
        self.__balance += amount  # Corrected to +=
        print(f"Deposited: {amount}")

    def withdraw(self, amount):
        if amount > self.__balance: # Corrected logic: if you ask for more than you have
            print("Insufficient balance!")
        else:
            self.__balance -= amount
            print(f"Withdrew: {amount}")

    def get_balance(self):
        return self.__balance

# Testing the code
b1 = Bank(100)
b1.deposit(5200)
print("Current Balance:", b1.get_balance()) # Added () to call the method

b1.withdraw(1000)
print("Final Balance:", b1.get_balance())
```
