```python
#Normally, methods need self to access an object's data (like a student's name). However, 
#a static method doesn't know anything about the object's specific data. Itâ€™s just a regular function that lives inside the class 
#because it is logically related to it.
```

```python
#Static Method
class Student:
    def __init__(self,name):
        self.name = name
    @staticmethod
    def hello(name):
        print("Welcome", name)
s1 = Student('Karan')
s1.hello(s1.name)
```
