```python
class School:
    def Teacher(self):
        pass
    def Student(self):
        pass

class classroom(School):
    def Teacher(self):
        print("Teacher is Teaching class")
    def Student(self):
        print("Students are Studying")

c1 = classroom()
c1.Teacher()
c1.Student()
```

```python
class Courses:
    def Course_info(self):
        a = input("Enter which course u want:")
    def duration(self):
        pass
class ExamInterfaces:
    def exam_type(self):
        pass
class Python_Course(Courses, ExamInterfaces):
    def duration(self, to):
        print("Duration is ", to)
    def exam_type(self, tr):
        print("Exam Type is : ", tr)

pc1 = Python_Course()
pc1.Course_info()
pc1.duration("5 Months")
pc1.exam_type("Practical")
```
